In this project, we will working with data name "laptops.csv" that contains 1,300 laptop computer and do some cleaning data:

```python
#Read file and import libraries
import pandas as pd
laptops=pd.read_csv("laptops.csv", encoding="latin_1")
laptops
```

# Cleaning columns name

```python
laptops.columns
```

```python
def clean_col(col):
    col = col.strip()
    col = col.replace("Operating System", "os")
    col = col.replace(" ","_")
    col = col.replace("(","")
    col = col.replace(")","")
    col = col.lower()
    return col
new_columns=[]
for i in laptops.columns:
    new_columns.append(clean_col(i))
laptops.columns=new_columns
laptops.columns
laptops
    
```

# Convert string columns to Numeric data

As the data shows that 13 columns's type is object, however there are some data must be convert to numeric:
- Screen size
- Ram
- Storage
- Weight
- Price

```python
#Clean format the screen_size column
laptops["screen_size"].unique()
laptops["screen_size"]=laptops["screen_size"].str.replace('"','')
laptops["screen_size"]=laptops["screen_size"].astype(float)
laptops.rename(columns={"screen_size":"screen_size_inch"}, inplace=True)
laptops["screen_size_inch"]
```

```python
#Clean the ram column
laptops["ram"]=laptops["ram"].str.replace("GB","")
laptops["ram"]=laptops["ram"].astype(int)
laptops.rename(columns={"ram":"ram_gb"},inplace=True)
```

```python
laptops["storage"].unique()

```

```python
laptops["weight"].unique()
laptops["weight"]=laptops["weight"].str.replace("kgs","").str.replace("kg","").astype(float)
laptops["weight"].unique()
laptops.rename(columns={"weight":"weight_kg"}, inplace=True)
```

```python
laptops["price_euros"]=laptops["price_euros"].str.replace(",",".").astype(float)
laptops["price_euros"]
```

# Extract data and create new columns

```python
#Extract data
laptops["gpu"].unique()
laptops["gpu_manufacturer"]=laptops["gpu"].str.split().str[0] #Extract GPU manufacturer name
laptops["cpu"].unique()
laptops["cpu_manufacturer"]=laptops["cpu"].str.split().str[0] #Extract CPU manufacturer
laptops
```

```python
laptops["cpu"].unique()
```

```python
laptops["processor_speed_GHz"]=laptops["cpu"].str.split().str[-1]
laptops["processor_speed_GHz"]=laptops["processor_speed_GHz"].str.replace("GHz","").astype(float)
laptops["processor_speed_GHz"]
```

```python
laptops["screen_resolution"]=laptops["screen"].str.split().str[-1]
```

```python
laptops["os"].unique()
```

There is duplicate value "Mac OS" and "macOS" that we should format

```python
mapping_dict={"Windows": "Windows",
              "No OS": "No OS",
              "Linux": "Linux",
              "Chrome OS":"Chrome OS",
              "macOS": "Mac OS",
              "Mac OS" : "Mac OS",
              "Android": "Android"}
laptops["os"]=laptops["os"].map(mapping_dict)
laptops["os"].unique()
```

# Dealing with missing data

As the os column and os_version column is closely related, we should take a look on this column before we choose a method to deal with missing data

```python
laptops.isnull().sum()
```

it depends on the purpose of using data we can leave null value as it is or cleanning it by dropping or replace new value.

```python
laptops["os_version"].value_counts()
```

```python
laptops["os_version"].unique()
```

- The "No OS" version must be understood is unknown version, so if the os_version is null and "os" is "No OS", we should replace the value of null by "Unknown Version"
- The Mac OS is the X version according to Wikipedia: https://en.wikipedia.org/wiki/MacOS so we should replace with version X

```python
laptops.loc[laptops["os_version"].isnull(), "os"].value_counts()
```

```python
laptops.loc[laptops["os"]=="No OS", "os_version"]="Unknown Version" #Replace the "No OS" value to "Unknown version"
laptops.loc[laptops["os"]=="Mac OS", "os_version"]="X" #To replace the "Mac OS" with "X" version
```

```python
laptops.loc[laptops["os_version"].isnull(), "os"].value_counts()
```

# Answering some business questions

<b> <red> Are laptops made by Apple more expensive than those made by other manufacturers? </red></b>

```python
manufacturer_price=laptops.groupby(["manufacturer"])["price_euros"].mean()
manufacturer_price.sort_values()
```

Compare to laptops from other manufacturer, Apple is quite expensive but it still cheaper than Microsoft, Google, MSI, LG, Razer

<b> What is the best value laptop with a screen size of 15" or more? </b>

```python
fifteen_inch_or_more=laptops.loc[laptops["screen_size_inch"]>=15]
fifteen_inch_or_more.sort_values(by=["price_euros"])
```
